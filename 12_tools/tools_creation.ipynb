# Custom Tool

```python
# Step 1: Create a function

def multiply(a,b):
    """Multiply two numbers"""
    return a*b
```

```python
# Step 2: Add type hints

def multiply(a : int, b : int) -> int:
    """Multiply two numbers"""
    return a*b
```

```python
# Step 3: Add tool decorator

@tool
def multiply(a : int, b : int) -> int:
    """Multiply two numbers"""
    return a*b
```

```python
result = multiply.invoke({"a":10,"b":50})
```

```python
result
```

```python
multiply.name
multiply.description
multiply.args
```

```python
multiply.args_schema.model_json_schema()
```

# Advanced schema definition

```python
class multiplyInput(BaseModel):
    num1 : int = Field(description = "The first number to multiply")
    num2 : int = Field(description = "The second number to multiply")
```

```python
@tool("Multiplication", description="Method to perform multiplication of two numbers",args_schema=multiplyInput)
def multiply(num1 : int, num2 : int) -> int:
    """Multiply two numbers"""
    return num1*num2
```

```python
multiply.args_schema.model_json_schema()
```

```python
result = multiply.invoke({"num1":140,"num2":500})
```

```python
result
```

# ToolKit

```python
class multiplyInput(BaseModel):
    num1 : int = Field(description = "The first number to multiply")
    num2 : int = Field(description = "The second number to multiply")
```

```python
class AdditionInput(BaseModel):
    num1 : int = Field(description = "The first number to Add")
    num2 : int = Field(description = "The second number to Add")
```

```python
class SubtractionInput(BaseModel):
    num1 : int = Field(description = "The first number to Subtract")
    num2 : int = Field(description = "The second number to Subtract")
```

```python
@tool("Multiplication", description="Method to perform multiplication of two numbers",args_schema=multiplyInput)
def multiply(num1 : int, num2 : int) -> int:
    """ Multiply two numbers"""
    return num1*num2
```

```python
@tool("Addition",description="Method to Add two numbers",args_schema=AdditionInput)
def addition(num1 : int, num2 : int) -> int:
    """ Addition of two numbers"""
    return num1*num2
```

```python
@tool("Subtraction",description="Method to Subtract numbers",args_schema=SubtractionInput)
def subtraction(num1 : int, num2 : int) -> int:
    """ Subtraction of two numbers"""
    return num1*num2
```

```python
class MathToolkit:
    def get_tools(self):
        return [multiply,addition,subtraction]
```

```python
toolkit = MathToolkit()
tools = toolkit.get_tools()
```

```python
for tool in tools:
    print(tool.name)
    print(tool.description)
    print(tool.args_schema.model_json_schema())
```
