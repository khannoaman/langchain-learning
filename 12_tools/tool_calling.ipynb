```python
from langchain_core.tools import tool
from pydantic import BaseModel, Field
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI
from dotenv import load_dotenv

load_dotenv()
```

# Tool Creation

```python
class multiplyInput(BaseModel):
    num1 : int = Field(description = "The first number to multiply")
    num2 : int = Field(description = "The second number to multiply")
```

```python
@tool("Multiplication", description="Method to perform multiplication of two numbers",args_schema=multiplyInput)
def multiply(num1 : int, num2 : int) -> int:
    """Multiply two numbers"""
    return num1*num2
```

```python
multiply.name
```

```python
multiply.description
```

```python
multiply.args
```

# Tool Binding

```python
model = ChatGoogleGenerativeAI(model="gemini-2.5-flash") 
```

```python
model_with_tool = model.bind_tools([multiply])
```

# Tool Calling

```python
model_with_tool.invoke("Hello, How are you?")
```

```python
response = model_with_tool.invoke("Can you tell me how much is 345 times 90.")
```

```python
response
```

```python
response.tool_calls
```

# Tool Execution

```python
multiply.invoke(response.tool_calls[0]["args"])
```

```python
multiply.invoke(response.tool_calls[0])
```
